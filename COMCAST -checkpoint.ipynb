Q.Provide the trend chart for the number of complaints at monthly and daily granularity levels.


```python
data['Date_month_year']= pd.to_datetime(data['Date_month_year'])
data['Date_month_year'].dt.month
data['Date_month_year'].dt.day
data['month']=data['Date_month_year'].dt.month
data['day']=data['Date_month_year'].dt.day
```

```python
x=data["month"].value_counts()
month_trend = pd.DataFrame(x)
month_trend = month_trend.rename(columns = {"month":"frequency"})
month_trend["month"] = month_trend.index
month_trend= month_trend.sort_values("month")
month_trend.head(5)
```

```python
sns.barplot(x="month", y="frequency",data=month_trend)
```

Q.Which complaint types are maximum i.e., around internet, network issues, or across any other domains.

```python
a = data['Customer Complaint'].value_counts()

complaint_types = pd.DataFrame(a)
complaint_types = complaint_types.rename(columns = {"Customer Complaint":"frequency"})
complaint_types["complaint_types"] = complaint_types.index
complaint_types.sort_values("frequency",ascending = False)
complaint_types.reset_index(drop=True,inplace= True)
complaint_types.head(10)
```

Q.Create a new categorical variable with value as Open and Closed. Open & Pending is to be categorized as Open and Closed & Solved is to be categorized as Closed.

```python
data["new_status"]=data["Status"].replace({"Solved":"Closed","Pending":"Open"})
data.head()
```

Q.Which state has the maximum complaints

```python
c = data['State'].value_counts()
State_count = pd.DataFrame(c)
State_count= State_count.rename(columns = {"State":"frequency"})
State_count["State"] = State_count.index
State_count.reset_index(drop=True,inplace = True)
State_count.sort_values("frequency",ascending = False,inplace = True)
State_count.head()
```

Georgia has the highest number of complaints

Which state has the highest percentage of unresolved complaints

```python
open_complaints = data[data['new_status']=='Open']
unresolved_complaints = open_complaints['State'].value_counts()
unresolved_comp= pd.DataFrame(unresolved_complaints)
unresolved_comp =unresolved_comp.rename(columns = {"State":"Frequency"})
unresolved_comp["state"] = unresolved_comp.index
unresolved_comp["percent"] = (unresolved_comp['Frequency'] / unresolved_comp['Frequency'].sum()) * 100
unresolved_comp.reset_index(drop=True,inplace= True)
unresolved_comp.head()
```

```python
plt.pie(unresolved_comp["Frequency"],labels = unresolved_comp['state'],radius= 3,autopct='%1.1f%%')
plt.show
```

Georgia has the highest percentage of unresolved complaints(15.5%)

Q. Provide the percentage of complaints resolved till date, which were received through the Internet and customer care calls.

```python
closed_complaints = data[data['new_status']=='Closed']
recieved_via=closed_complaints [closed_complaints ["Received Via"].isin(["Internet","Customer Care Call"])]
recieved_via = closed_complaints["Received Via"].value_counts()
recieved = pd.DataFrame(recieved_via)
recieved = recieved.rename(columns = {"Received Via":"frequency"})
recieved["medium"] = recieved.index
recieved.head()
```

```python
plt.pie(recieved["frequency"],labels = recieved['medium'],radius= 1,autopct='%1.1f%%')
plt.show
```

Q.Provide state wise status of complaints in a stacked bar chart

```python
x = data[data['new_status']=='Closed']
ccs = x['State'].value_counts()
cc = pd.DataFrame(ccs)
y = data[data["new_status"]=="Open"]
ocs= y['State'].value_counts()
oc = pd.DataFrame(ocs)
```

```python
final= pd.concat([cc,oc],axis=1)
final.columns.values[0] = "open"
final.columns.values[1] = "close"
final["state"] = final.index
final.reset_index(drop=True,inplace= True)
final.head()
```

```python
final.plot(x="state",kind="bar",stacked=True,figsize=(10,10))
plt.ylabel("no of complaints")
plt.title("stacked barplot of complaints in each countries")
plt.show()
```
